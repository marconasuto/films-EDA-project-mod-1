### Web-scraping

```python
def parse_content(url):
    '''This function parses an html page, given its url, and returns a BeautifulSoup object.'''
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup
```

```python
#Get titles and linkpages to (ultimately) imdbpro from boxofficemojo.com


#urls we're interested into scraping taken from ranking list on boxofficemojo.com
years = ['2014','2015','2016','2017','2018','2019','2020']
url = 'https://www.boxofficemojo.com/year/world/'
urls = [url + '{}'.format(x) + '/' for x in years]

#parse with Beautiful Soup, get dictionary containing soup objects of the ranking lists of films in the last 3 years, 
#ordered by box office

soup = {}
for url,year in zip(urls,years):
    soup[year] = parse_content(url)

#get, through looping, a dictionary containing soup objects for each url, in the format of
#<https://www.boxofficemojo.com/releasegroup><film_id>
#from this dictionary, get the links to each film's profile page on www.imdbpro.com
  
container = {}
titles = {}
linkpages_title = {}
urls_mojo = {}
soup_mojo = {}
url_root_mojo = 'https://www.boxofficemojo.com'

for year in years:
    container4link = {}
    container[year] = soup[year].findAll('a', href=True)
    titles[year] = [x.text for x in container[year] if re.search('^/releasegroup',x['href'])]
    linkpages_title[year] = [x['href'] for x in container[year] if re.search('^/releasegroup',x['href'])]
    urls_mojo[year] = [url_root_mojo + x for x in linkpages_title[year]]
    print('Year:{}'.format(year))
    for url in enumerate(urls_mojo[year]):
        container4link[titles[year][url[0]]] = parse_content(url[1])
        print('Title {}'.format(titles[year][url[0]]), '{} films left for {}'.format(len(urls_mojo[year])- url[0], year))
    soup_mojo[year] = container4link
```

```python
#for safety, copy of BeatifulSoup object
soup_mojo_copy = soup_mojo.copy()
```

```python
#get urls to cast for each film on www.imdbpro.com
singlesoup = {}
url_pro = {}
for year, names in titles.items():
    linkscast = {}
    for name in enumerate(names): 
        singlesoup = soup_mojo[year][name[1]]
        linkscast[name[1]] = singlesoup.findAll('div',class_='a-box-inner')[0].a['href']
    url_pro[year] = linkscast 
```

```python
#for safety, copy of the url dictionary
url_pro_copy = url_pro.copy()
```

```python
#get BeautifulSoup objects for each film and store them in a dictionary
soup_pro = {}
for year in years:
    container4link = {}
    for name,url in url_pro[year].items():
        container4link[name] = parse_content(url)
        print('Title {}'.format(name), year)
    soup_pro[year] = container4link
```

```python
#from BeautifulSoup objects, get containers for each film so to get cast, duration, genre, budget and boxoffice data 
container_cast = {}
for year in years:
    container4title = {}
    for name, container in soup_pro[year].items():
        container4title[name] = container.findAll('a',class_='a-size- a-align- a-link-')
    container_cast[year] = container4title
```

```python
#get actors, writers, producers, genre, certificate, duration, budget, boxoffice in the us and worlwide
#handles to use to get names of the first 10 actors
handles = []
for i in range(0,10):
    handles.append('cst_' + str(i+1) + '$')

#initialize dictionaries
film_dict = {}
films = {}

#imdb url
imdb_url = 'https://www.imdb.com/title/'

#patterns to retrieve budget, boxoffice
pattern_us = re.compile('Gross USA')
pattern_ww = re.compile('\nCumulative Worldwide Gross')
pattern_bdg = re.compile('\Budget')

for year in years:
    film_dict = {}
    for name, container in container_cast[year].items():
        temp_dict = {}
        temp_dict['directors'] =  [x.span.text for x in container if re.search('dir$',x['href'])]
        temp_dict['actors'] = [x.text for x in container for el in handles if re.search(el, x['href'])]
        temp_dict['writers'] = [x.span.text for x in container if re.search('wrt$',x['href'])]
        temp_dict['producers'] = [x.span.text for x in container if re.search('prodr$',x['href'])]
        temp_dict['year'] = year
        
        try:
            temp_dict['genres'] = soup_pro[year][name].findAll('span', {'id':'genres'})[0].text
            temp_dict['certificates'] = soup_pro[year][name].findAll('span',{'id':'certificate'})[0].text
            temp_dict['duration'] = soup_pro[year][name].findAll('span',{'id':'running_time'})[0].text
        except: 
            temp_dict['genres'] = None
            temp_dict['certificates'] = None
            temp_dict['duration'] = None
        
        #get from ulrs for each title on imdb budget, boxoffice (gross US and worldwide)
        url_film = imdb_url + url_pro[year][name].split('/')[4]
        print(url_film)
        temp_soup = parse_content(url_film)
        div_cont = temp_soup.findAll('div', class_='txt-block')
        
        for el in enumerate(div_cont):
            if (re.search(pattern_us, el[1].text)):
                try:
                    temp_dict['Gross US'] = float(el[1].text.split('Gross USA: $')[1].split('\n')[0].replace(',',''))
                except: 
                    temp_dict['Gross US'] = None
            elif (re.search(pattern_ww, el[1].text)):
                try:
                    temp_dict['Gross Worldwide'] = float(el[1].text.split('Cumulative Worldwide Gross: $')[1].split('\n')[0].replace(',','')) 
                except:
                    temp_dict['Gross Worldwide'] = None
            elif (re.search(pattern_bdg, el[1].text)):
                try:
                    temp_dict['Budget'] = float(el[1].text.split('Budget:$')[1].split('\n')[0].replace(',',''))
                except:
                    temp_dict['Budget'] = None
        film_dict[name] = temp_dict
    films[year] = film_dict
```

```python
#creating dataframes 
df2020 = pd.DataFrame(films['2020']).transpose()
df2019 = pd.DataFrame(films['2019']).transpose()
df2018 = pd.DataFrame(films['2018']).transpose()
df2017 = pd.DataFrame(films['2017']).transpose()
df2016 = pd.DataFrame(films['2016']).transpose()
df2015 = pd.DataFrame(films['2015']).transpose()
df2014 = pd.DataFrame(films['2014']).transpose()
```

```python
#concatenating, making titles a column, resetting index, dropping column index
scraped = pd.concat([df2020, df2019, df2018, df2017, df2016, df2015, df2014])
scraped['titles'] = scraped.index
scraped.reset_index(inplace=True)
scraped.drop('index',axis=1,inplace=True)
scraped
```

```python
#exporting to csv
scraped.to_csv(r'films_scraped.csv', index = False)
```
